### What is machine learning?

- The field of Machine Learning (ML) seeks to answer the question "How can we build systems that improve with data, and what are the fundamental laws that govern learning processes?"





### What do we mean by **machine** and **learning**?

- We need an algorithm with instructions to create a model that learns from data.

- To make accurate predictions, a model requires data and informative features.
- Complex machine learning models are used to identify relationship between features, and a desired output.

### The brief history of Machine Learning (ML)

- The field of machine learning has its roots in the 1950s, with early research on artificial intelligence and pattern recognition. 

- Pioneers in the field include Alan Turing, Claude Shannon, and Norbert Wiener. 

- Today, some notable pioneers in the field of machine learning include Geoffrey Hinton, Yann LeCun, and Yoshua Bengio, who have made significant contributions to the development of deep learning.

### Working with data

- There are a number of overlapping fields and approaches that deal with data.
- Example: Dataset with house properties and prices
    - **Data science**: Create and maintain a database and a dashboard to access house price data. 
    - **Computer science**: Create a search engine for the database.
    - **Statistics**: Test if a neighbourhood has a significant impact on house prices.
    - **Machine learning**: Train a model that predicts the price of a new type of house (tiny home) entering the market.

- The different fields related to data are overlapping, with a focus on different aspects. 

- E.g. data science focuses on the pipeline of organizing data, dealing with large datasets in the cloud, and visualising relevant information.

- Statisticians focus on specific, well-defined properties of the data, such as hypothesis tests. Statistical models are usually simpler than in ML, with a focus on interpretability.

- ML blends the principles of computer science and statistics to create sophisticated algorithms. ML can be used to:
  - generate new data (generative learning), 
  - identify groupings (clustering), 
  - predict new values for unseen observations (regression), 
  - predict labels for unseen observations(classification), 
  - complete a task (reinforcement learning),
  - or make recommendations (information filtering).

- The common theme for ML is automation. We want to know how we can automate processes (prediction, clustering, generation, recommending, task completion) by learning directly from data.

### Supervised learning vs unsupervised learning**:

**Supervised learning** involves training an algorithm on a labeled dataset, and the goal is to make predictions based on input features, with the help of labeled examples. The labeled examples provide the algorithm with information on the correct output for a given input.

e.g. Regression (prediction of continuous numerical values) and classification (prediction of a categorical class label).

**Unsupervised learning** is a type of machine learning where the algorithm is trained on an unlabeled dataset and tries to identify patterns or relationships in the data without any prior knowledge of the correct output. The goal is to understand the underlying structure of the data.

e.g. clustering (grouping objects based on the similarity of their attributes), self-supervised learning (e.g. predicting masked words in text to build language models).

**Other ML approaches**: reinforcement learning (learning with a reward, action and value space).

#### Deep Learning

Deep learning has seen succesful developments and applications in the past year. Deep learning defers from other "traditional" ML methods (such as the perceptron, logistic classification, linear regression, etc.) on few aspects including:

- **Architecture**: traditional ML algorithms have a shallow architecture (i.e., one or a few hidden layers) while deep learning algorithms have a deep architecture (i.e., many hidden layers).

- **Feature Engineering**: Some traditional ML algorithms necessitate extensive feature engineering, where relevant features need to be extracted from raw data. Deep learning algorithms, on the other hand, can automatically learn useful features from data, without manual selection of inout features or an exploratory data analyis.

- **Large Amount of Data**: Deep learning algorithms perform better with larger amounts of data, which might not be the case with more traditional ML algorithms.

- **Non-linear Relationships**: Deep learning algorithms model non-linear relationships between inputs and outputs, while traditional ML algorithms might not include non linear transformations of the input to draw non linerar relationships with the output.

- **Computational Requirements**: Deep learning algorithms are computationally intensive, requiring powerful GPUs for training, while traditional ML algorithms are typically less computationally demanding.

In summary, classical ML algorithms are suitable for smaller, well-defined problems with a limited amount of data, while deep learning algorithms are well suited for more complex, large-scale problems with unstructured data.

### ML pipeline

### Is the full ML pipeline needed to solve business problems?

- ML can be used to automate a task or a decision making process or to add new information to existing data.
- For data analysis, Exploratory Data Analysis (EDA) methods or data visualisation can be sufficient.
- Depending on the nature of the problem, a traditional statistical or rule-based approach may be sufficient. 

One example of a business problem that can be solved with EDA and does not require machine learning is determining the factors which influence customer satisfaction. 

- This can be done through analysing customer survey data and identifying patterns and correlations between different survey questions. 
- Techniques such as frequency counts, cross-tabulations, and chi-squared tests can be used to identify relationships between variables. 
- For example, you might find that customers who rate their overall satisfaction highly also tend to rate the product quality and customer service highly. 
- This type of analysis can provide valuable insights into what factors are driving customer satisfaction and help a business make decisions on how to improve it.

### The two broad categories of ML

- Supervised and unsupervised learning

### Supervised Learning

- Data includes **input** features $X$ and an **output** $Y$ (also called observation).
- Develop a model which generates the process $f$ such that $Y = f(X)$.
- The goal of a supervised learning model is to approximate the data generation function to make predictions $\hat{Y}$ with:

    $\hat{Y} = f'(X)$
   

Note that $f(X)$ is a stochastic process - such that $Y$ is drawn from a distribution.

Examples of a supervised learning problem:

- **Predicting house prices**:
    - Features could include: number of rooms, postcode, m$^2$ $\Rightarrow$ Prediction: distribution of house prices
- **Predicting whether a person has diabetes**:
    - Features could include: glucose concentration, blood pressure, age $\Rightarrow$ Prediction: probability of having diabetes

### Modeling steps: supervised example

- Example of modeling steps for a supervised **prediction** task:
  - **Define quantitative target**: minimise mean squared error
  - **Choose model**: neural network model
  - **Train model**: gradient descent
  - **Evaluate model**: test predictions on novel data

### Unsupervised Learning

- Find structure in the **data** $X$
- **Group** similar data points (clustering)
    - e.g. grouping similar users in social networks
- Learn to **generate** data: $P(X)$
    - e.g. generate novel text for a topic
- Reduce the number of **data** dimensions
    - e.g. find informative dimensions for 2D visualisation

### Modeling steps: unsupervised example

- Example of modeling steps for an unsupervised **clustering** task:
    - **Define quantitative target**: likelihood of the data for generative model $\hat{P}(X)$
    - **Choose model**: Gaussian mixture model
    - **Train model**: expectation-maximisation
    - **Evaluate model**: compare clusters with ground truth

A large part of these steps is automated with libraries such as scikit-learn (`sklearn`), and the main task in modeling is choosing a sensible model and its settings and hyperparameters. This is in addition to preprocessing the data.

### Choosing a model

- There are many factors that indicate the most appropriate model:
    - **Data type**: tabular, image, time series, text
    - **Data properties**: # dimensions, # samples, missing values
    - **Task**: prediction, clustering, visualisation, control
    - **High-level goals**: interpretability, performance, robustness, computational load

An important part of modeling expertise is to be familiar with the broad range of ML models, and which to choose for each problem. There is a large number of available models, each with their strengths and weaknesses. In this course we cover many models, with examples of when they can be useful.

### Model zoo





For each task and type of data, there are multiple model choices. Part of ML expertise is knowing which to use and how to tune its hyperparameters and train it. Libraries make it easy to try multiple models with your data.

### What to achieve from the data? Generalisation!

- In supervised learning we want to use the available data to:
    - (1) learn a predictive model (e.g. weight vector)
    - (2) assess the model's performance
    
- We train the model on a training dataset and assess the model empirically on a test set. We split the data into two parts:
    - Data for (1): **training set**
    - Data for (2): **test set**
    
- In addition to the training and test set, when the data is large enough, it is recommended to use a validation set to finetune a model's hyperparameters and avoid overfitting on the test set (more on this later).
   

### How does a model generalise to new data?
    
- A model usually performs well on the training data, like **curve-fitting** with many parameters.
- A fundamental concept in machine learning is that a model should perform well on **unseen data**, i.e. the test set.
- A model that performs well on the train set does not mean it also performs well on unseen data.
- Models have different levels of **complexity**, and should be chosen according to the amount ad type of available data. 
- **Bias** is the simplifying assumptions made by the model to make the target function easier to approximate. 
- **Variance** is the amount by which the estimate of the target function changes given different training datapoints.
- **Underfit (figure on the left)** - High Bias, low Variance
- **Overfit (figure on the right)** - Low Bias, high Variance
- There is always a trade-off between bias and variance
