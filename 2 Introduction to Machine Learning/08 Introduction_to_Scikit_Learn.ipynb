## Supervised and unsupervised learning in scikit-learn

Machine learning is a name for a set of techniques that are designed to automatically learn patterns and insights from data without being explicitly programmed to do so. It works by using statistical algorithms to analyze large sets of data, identifying patterns and relationships between different variables.

It's a powerful tool that can help us make sense of complex data and solve problems more efficiently.

Classical machine learning problems can be largely split into two different categories:

1.  Supervised Learning (SL) $\rightarrow$ predicting a *target* (e.g. a label, multiple labels, a time series, or a conitnuous variable) from one or more *features* in the dataset. Each data point in your dataset consists of one or more features assigned to it and an associated target. Supervised learning learns a model which maps each set of features (input) to a target (output). If the target is a continuous value, the problem is called *regression*. If it is a discrete class, the problem is called *classification*. 
   
<img src = 'img/supervised-machine-learning.png'/>

1.  Unsupervised Learning (USL) $\rightarrow$ finding *patterns* in *unlabelled* data. We typically consider *clustering* problems where each data point is assigned to a different *cluster* or latent grouping of the data (though there are other types of unsupervised learning, such as dimensionality reduction!). The primary concept of unsupervised learning is that the algorithm generates new representations of the input data without the need of human intervention.



### Data

* Supervised learning requires *labelled* input and validation data, so that the algorithm can learn the relationships between the input features and the *ground-truth* target.
* Unsupervised learning can make use of *unlabelled* data, as the algorithm generates new representations of the data based on its underlying distribution (e.g. discrete clusters, new features)

## The Scikit learn way of doing things

When working with sklearn, we follow a standard procedure to operate the objects in the library, which follow a typical machine learning pipeline when training and evaluating a model.

A common approach is to apply the following steps:

    a) Data Preparation: Prepare and clean the data by handling missing values, encoding categorical variables, and splitting data into train and test sets.

    b) Choose a Model: Choose the appropriate algorithm for the problem, and create an instance of the model class.

    c) Training: Fit the model on the training data using the fit() method.

    d) Prediction: Use the model to make predictions on new data using the predict() method.

    e) Evaluation: Evaluate the performance of the model using appropriate metrics such as accuracy, precision, recall, and F1 score.

    f) Hyperparameter Tuning: Fine-tune the hyperparameters of the model using techniques such as grid search or random search to improve performance.

    g) Save and Load Model: Save the trained model to disk using the joblib library to reuse it in future by passing the file name in the dump() or load() functions.

The scikit-learn API is consistent across all algorithms, making it easy to switch between models and try different methods. The API follows a simple and concise interface for fitting, predicting, and evaluating models, making it accessible to users with a wide range of skill levels.

## sklearn classes

sklearn uses *Object oriented programming* concepts, and contains a variety of different *classes* that we can use for data processing, modelling, mining and so on. 

Classes in scikit-learn are used to define *objects* that have *attributes* (i.e. variables associated with the object) and methods (i.e. functions associated with the object). 

When using a class, you:
- create an instance of the class, specifying the *parameters* as arguments
- call its methods, possibly passing one or more arguments

This provides a higher level of abstraction and encapsulation, making it easier to organize your code and reuse parts of your machine learning pipeline.

Let's look at an example of using a class:

```python
# import LinearRegression class from linear model module
from sklearn.linear_model import LinearRegression 

# create an instance of the linear regression model, specifying that we do not want to fit an intercept
lr = LinearRegression(fit_intercept=False)

# use the get_params method of the linear regression object to retrieve the model parameters
lr.get_params()
```


Here are some example classes you might come across in the future:

| module            | Class              | Description                                                             | Example methods    | Example attributes    |
|-------------------|--------------------|-------------------------------------------------------------------------|--------------------|-----------------------|
| `linear_model`    | `LinearRegression` | Ordinary least squares Linear Regression.                               | `fit`, `predict`   | `coef_`, `intercept_` |
| `preprocessing`   | `StandardScaler`   | Standardize features by removing the mean and scaling to unit variance. | `fit`, `transform` | `mean_`, `var_`       |
| `model_selection` | `TimeSeriesSplit`  | Time Series cross-validator                                             | `split`            | None                  |


Classes are used to perform a range of tasks and the main difference between them is their level of  __abstraction__  and the control they provide when you use them. You can also make your own classes in sklearn!

#### Naming convention

How can we tell if we are importing a class or a function?

sklearn makes it easy to tell with a regular naming convention!
- Classes use `PascalCase`, where each word is capitalised and joined together with no spaces
- Functions use `snake_case`, where lower case is used throughout, and underscores are used to separate each word

## Abstraction in sklearn

sklearn uses the principle of *abstraction*
- we do not need to understand how the code works
- we only need to understand how to use it!

Think about the steering wheel in a car - it is straightforward to use, even though you don't necessarily know how it works!

Predictive models in sklearn use the same structure, e.g.:
- create an instance of the class, setting the parameters
- `fit` the model to some data
- use the model to `predict` some data
- `score` the model on some data

This makes it very easy to reuse our code to test multiple models!
- There are similar regular structures for transformers, cross-validation etc. 

#  An example - supervised regression

Let's look at an example, to show how abstraction makes it very easy to apply models and reuse code!

Don't worry about the theory here - the point is to show how easy it is to apply a new model that you have not necessarily seen before!

We start with a model you should have seen before: linear regression.

```python
from sklearn.linear_model import LinearRegression #import the linear regression estimator

lr = LinearRegression()
```

We also need some data to fit the model. Let's use the `make_regression` function we used before:

```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=1000, n_features=3, noise=100)
```

We use another function to split our data into train and test data (more on this later!)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

We can now fit our model to the training data:

```python
lr.fit(X_train, y_train)
```

And see how well the model predicts the training data:

```python
score = lr.score(X_test, y_test)

print(f'R2 for linear regression: {score:.3f}')
```

Great - so we have managed to generate data, fit a model, and assess its performance on out-of-sample data with only a few lines of code!

But what if we wanted to try a different model, for instance a Decision Tree?

Let's see how easy abstraction makes it!

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)

score = dtr.score(X_test, y_test)
print(f'R2 for decision tree regressor: {score:.3f}')
```

We used the decision tree in exactly the same way as the linear regression model.

In fact, you could try any regression model in sklearn using the same code.
