## Iteration - for loops

- Make sure you run the following code cell before you attempt any of the questions
- This will create `ages`, `names` and `details` variables. These will be referenced during following questions

````python
ages = [17, 22, 14, 55, 23, 19, 16, 21, 33, 42, 25]
names = ['John', 'Mandy', 'Siri', 'Robert', 'Saul', 'Paulina', 'Fay', 'Mahmood', 'Ben', 'Angela', 'Vicky']
details = list(zip(names, ages))
````

**Example1:** You have been given the below code that counts the number of elements in `ages` list. This is a step-by-step process where you:
- First create a counter, say variable `counter`, and set its value to `0` 
- Then use a `for` loop to iterate through the list `ages`, incrementing `counter` on each loop by `1`  
- At the end, `print` the statement **'The number of ages in the list is `<your result>`**, where `<your result>` is the new value of `counter`
- You can make use of `f-strings` or the `.format()` method while creating the `print()` statement

````python
counter = 0

for age in ages:
    counter = counter + 1

print(f'The number of ages in the list is {counter}')
````

**Example2:** You have been given the below code that use a `for` loop to iterate through each tuple in `details` list** 
- If the `age` value in tuple is less than `min_age`, print **'`<name>` is too young'
- `<name>` is the `name` value appearing in the tuple

````python
details
````

````python
min_age = 18

for person in details:
    if person[1] < min_age:
        print(f'{person[0]} is too young')
````

**Q3)** Copy your code from above into the cell below: 

- Create a variable, `number`, before the `for` loop and set its value to `0`
- Next, add a further statement to the `for` loop to count the number of people who are not too young (this is captured by the variable `number`)
- `print` a final statement **'The number of people who can attend is `<number>`'**

**Example3:** You have been given the below code that prints out a single list of those who are too young and those who aren't. This is a step-by-step process where you: 

- Create two empty lists called `refused` and `inside` respectively
- Then loop through `details` and use `.append()` to add the person's `name` to `refused` list if they are too young, and to `inside` list otherwise. (See the Introduction to Python lessons if you want to refresh your memory of `.append()`)
- `Print` out both lists

````python
refused = []
inside = []

for person in details:
    if person[1] < min_age:
        refused.append(person[0])
    else:
        inside.append(person[0])

print(inside)
print(refused)
````

**Q4)** `len()` function can be used to calculate the number of people in the  `inside` and `refused` lists.
