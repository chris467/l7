# Part 1 - Introduction to Pandas

## ☑️ What is Pandas?

- A Python library for **data manipulation and analysis**.

- Simplifies common data analysis tasks:
    - **Load data** from numerous type of files and online sources
    - **Fast and efficient** handling of large amounts of data
    - **Filtering**, **sorting**, **editing** and **processing** of data
    - **Joining** and **aggregation** of datasets

- Tools for **time series** and **statistical analysis**
- Display of data in **tables** and **charts**   
- Provides two valuable data structures: **DataFrame** and **Series**

## ☑️ Pandas Data Structures
- Data structures are used to **store** and **organise** data.

#### DataFrame
- Pandas `DataFrame` is a **2-dimensional** labelled data structure made up of **rows** and **columns** of potentially different data types
- You can think of it like a spreadsheet or SQL table. It is generally the most commonly used data structure
- Indexing in a `DataFrame` starts from **0** (for both rows and columns)

#### Series
- Pandas `Series` is a **one-dimensional** labelled data
- An example of a `Series` is one column from a DataFrame
- Indexing in a `Series` starts from **0**

## ☑️ Retrieving data from a DataFrame

- This is known also as **indexing**. There are two ways to locate rows and columns in pandas using:
    - **positions**: zero-based integers for row  and column location
    - **labels**: row labels (numeric by default) and column labels/names

#### The DataFrame Index

- The row labels (which can be numeric or text-based) are known as the **index** of the DataFrame
- When a DataFrame is created, Pandas by default creates an integer-based index, called a **RangeIndex**
- The `.index` attribute has information about the index of the DataFrame

````python
df.index
````

### Use `.iloc[]` for position-based indexing



````python
df.iloc[0:3, 0:5]
````

````python
df.iloc[0:3, [0,1,9]]
````

- Note the use of **[brackets]** rather than (parentheses)
- The syntax is `[rows,columns]`, where `rows` and `columns` can be **either single values or ranges**
- Ranges work in the same manner as Python lists (0:3 returns three elements, from positions 0, 1 and 2)

````python
df.iloc[0:3, :]
````

````python
df.iloc[0:3]
````

- If only a single value or single range is provided, all columns for the given rows will be returned (unless preceeded by `:,`, in which case all rows for the given column(s) would be returned)

````python
df.iloc[8000:8002]
````

- If the index is not found or out of range, an empty DataFrame will be returned (rather than an error message)

````python
df.iloc[0]
````

- Explicitly requesting a specific single row or column will return a pandas Series rather than a DataFrame

### Use `.loc[]` for label-based indexing



````python
df.loc[0:3, ['customerid', 'tenure', 'churn']]
````

- In this instance we could have used `.iloc` instead of `.loc`, because the position-based and label-based row indexes are equivalent
- If the row index was text-based (for example, passenger names), we would need to use `.loc`

````python
df.loc[:, 'customerid']
````

- As with `.iloc()`, we can provide a pair of values (or a pair of groups of values), in the format `[rows,columns]` 
- Providing only a single value or value group will result in indexing only by row, with all columns being returned
- Groups of values can be provided as a [list], (tuple), or {set}; if a set is provided, the original column order will be retained in the output

- Note how `0:3` is interpreted differently here, in that this refers to the **labels** of the row index rather than the **position**, and that row `3` is included, whereas in the positional syntax (and usual Python list syntax) this would not be the case.

- **label-based indexing** (`.loc`) is **more robust** for looking up data, in that changes to the order or presence of rows or columns will not result in the wrong data being returned

## ☑️ Updating Values Found Using Indexing

- We can use these indexing methods to both **get** (i.e. retrieve values) and **set** (i.e. add or modify values)

#### An example

````python
df.head()
````

````python
new_df = df.copy()

new_df.loc[0, 'paperlessbilling'] = 'No'

new_df.head()
````

```python
new_df = df.copy()
```
- We created a **copy** of the original DataFrame `df`, so that our modifications do not affect it

```python
new_df.loc[0, 'paperlessbilling'] = 'No'
```

- We used label-based indexing to locate a specific **cell** and assign the value `No` to it

- Using `new_df = df` would have meant that subsequent modifications made to the data in `new_df` also affected `df`
- This is due to the way Pandas works, sometimes creating a **view** of the original DataFrame rather than a **copy** of it
- The rules behind when and why pandas creates views or copies are quite complex; if in doubt, consider using `.copy()`

### Row selection with multiple conditions

You can create complex conditions by combining logical operations with the following python bitwise operators:

| Operator | Description   |
|:- |:- |

|   **OR (\|)**  | check both expressions are logically True|

#### An example filtering with multiple conditions - AND (&)
- Question - How many customers have signed up for phone service **and** have a tenure of more than 24 months?

### Setting the index using values from a column

- Provide a faster way to lookup data 
- Index values **do not have to be unique** and can be **numbers or text**
- The number of rows and row order remain unchanged
- Data columns can be used as the index and vice-versa

````python
df2 = df.set_index('customerid')
df2.head(3)
````

````python
df3 = df.set_index('customerid', drop=False)
df3.head(3)
````

````python
df2.index
````

- The `set_index` method can be used to set a given column as the index
- By default, the chosen column is dropped from the DataFrame 
    - Here we have overwritten the default parameter using `drop=False`; we see the `customerid` values both as the index and as a column within the DataFrame itself

- Note how on each occasion we assigned the returned DataFrame to a new variable (`df2` or `df3`)  
    - When using the default parameters for the `.set_index` method, the original DataFrame `df` is not modified

- We can assign the result to a new variable, or we can use the `inplace=True` parameter when calling the `set_index` method, which will modify the original `df` DataFrame

````python
df2.reset_index(inplace=True)
df2.head()
````

- If we have an index which we would prefer to have as a normal data column, the `reset_index` method can be used
- Again use `inplace=True` if we want to modify the existing DataFrame

### Removing columns with `.drop()`

````python
df_new = df.drop(['phoneservice', 'internetservice'], axis=1)
df_new.head(3)
````

- The `.drop()` method allows us to specify labels of rows or columns to drop
- The `axis` parameter determines whether the labels refer to rows (`0`, the default) or columns (`1`)
- We have assigned the result to a new variable
- We could have used `inplace=True` on the original DataFrame, although this would be **destructive** in that the columns would be lost, and we would need to re-run all of our code in order to retrieve the lost columns

### Convert the data type with `.astype()`

### Applying a function to modify data with `.apply()`

````python
def phone_bool(text):
    
    if text == 'Yes':
        return True
    else:
        return False
````

- In the above example, we have defined a custom function `phone_bool` that convert `Yes` and `No` values to `True` and `False`. Not to worry you will learn more about custom functions and if/else statements in the next module

````python
df.head()
````

````python
df['phoneservice'] = df['phoneservice'].apply(phone_bool)
df.head()
````

- We used the `.apply()` method to call the `phone_bool` function with the value in the `phoneservice` column as the argument
