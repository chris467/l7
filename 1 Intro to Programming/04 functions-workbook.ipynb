**Example1:** You have been given the below code.

This section illustrates how a user may try to use a function in an unintended way. To overcome such situations you have been given the below code for the function `error_catching`. As you can see this function will display **'Only numbers should be used. Please try again.'** if any of the parameters is a string.

```python
def error_catching(a, b, c):
    if (type(a) == str) or (type(b) == str) or (type(c) == str):
        result = 'Only numbers should be used. Please try again.'
    else:
        result = (a + b) * c
    
    return result
```

```python
print(error_catching(5, 6, 'seven'))
```

**Example3:** You have been given the below code for the function called `names_dict`, which takes a `list` and returns a `dictionary` where:
- There is a `key` for each unique first character found in the list.   
- The `values` are a list of the entries which begin with the character used for the `key`.
- The lists of entries are ordered alphabetically.

We will build up `names_dict` function step-wise below. The below function takes a list of strings (eg, names) and returns a list of unique, first characters of those names, `keys`.  Test it using the list provided.

```python
def names_dict(names):
    first_chars = []
    
    for name in names:
        first_chars.append(name[0])
    
    keys = set(first_chars)
    keys = sorted(keys)
    
    return keys
```

```python
list(names_dict(names_list))
```
