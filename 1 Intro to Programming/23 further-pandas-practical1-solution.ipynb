# Practical 1 - pivot, pivot_table, melt

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/countries.csv', decimal=",")
df.head()
```

Note that there are some missing values - we'll need to keep that in mind during our analysis.  

**Exercise 1** Create a `pivot_table` which shows the `sum` of `Population` and `Area (sq. mi.)` for each `Region`.

```python
pd.pivot_table(df, values=['Population', 'Area (sq. mi.)'], index='Region', aggfunc='sum')
```

**Exercise 2** Use `melt` to create a DataFrame where all of this information is shown in three columns, called `Country`, `Metric` (containing the variable name) and `Measurement` (containing the value for that given variable for the given country). The resulting DataFrame should have 4313 rows.

```python
melted = pd.melt(df, id_vars='Country', var_name='Metric', value_name='Measurement')
melted.shape
```

**Exercise 3** Create a subset of the melted DataFrame containing only data for Afghanistan, with the `Country` column removed and `Metric` as the index (using `.set_index('Metric')`).

*Hint: examine the country values if you are having trouble filtering the DataFrame.*

```python
afgh = melted[melted['Country'].str.contains('Afghanistan')].copy()
# Copy is needed because a view reference is returned instead of the actual dataframe. 
# This makes any subsequent transformations non-determinate whether they change teh view or the dataframe
# https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy

afgh.drop('Country', axis=1, inplace=True)

afgh = afgh.set_index('Metric')
afgh
```
