**Exercise 9**: You can also include slices with a *step*. The syntax is: `a[start, stop, step]`. 
For instance, 
```
a = np.array([1,2,3,4,5,6])
a[2:6:3]
```
Will get the range `2:6` and return every 3rd item. Try it out.

**Exercise 10**: Another useful operator is `::`. Try out `a[::2]` to see what it does. 

With the operator `::`, we used the syntax introduced in the previous exercise, with:
- start not specified, so it defaults to the beginning of the array.
- stop not specified, so it defaults to the end of the array.
- step: The 2 in ::2 specifies the step value, which means that elements are selected with a step of 2. This means it will include every second element from the original array a, starting from the first element.
