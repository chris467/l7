## Motivation for version control

##### What is version control and Git?
- Version control is a system to record changes to a file, or set of files, over time to recall and monitor every update to the codebase.
- Git is a version control system (VCS) for tracking all your code changes. 
- Git is a _distributed_ type of VCS. You might have used Subversion before which is a _centralized_ VCS.
- Although other VCS systems are still in use, Git has become the de facto standard VCS for the industry
- Read more about the fundamentals of version control in the [Git documentation](https://git-scm.com/book/en/v2/Getting-Started-About-Version-Control).


In general, version control is a system for _tracking changes in files_. Any time a line of code is added, modified or removed, it is considered a change. The same is true for adding and removing files. We need a version control system within software development because it is beneficial to track how code updates, improves and changes over time with some respective justification for each change. With many people working on a project at once, we need to ensure synchronisation and progressive development so that everyone's changes don't overwrite each other. 

Git provides all of this functionality through the `git` command line tool, and technically that is enough to run a completely decentralized, distributed version control system. To make developers' life easier, remote Git platforms such as GitLab, GitHub or BitBucket exist to provide some centralization and better user experience working with Git. Git is a _distributed_ version control system which means you and the remote server both have complete versions of the codebase which you keep synchronised between changes.

##### Why do I need it?
- You're **always** working in a team:
    - Share code with your **co-workers**.
    - **Future you** will not remember why **past you** made that change!
- Save and commit often:
    - Reduces anxiety working on a big project.
    - Every code change is recorded with explanation messages.
    - Keeps a clean top level directory with entire history retained. 
- Easy code sharing and switching between code changes in different branches.

In short, you need version control because you will **undoubtedly** forget what that function does when you need it in 2 years. Git encourages development in packages of **dedicated functionality changes** which is useful when you want your colleagues to use your code. But this is also useful for your own work as every decision you made is **recorded and explained** for the future. 

Mastering how to make contributions to a code project using Git opens up many possibilities. At the minimum, it helps your employability by demonstrating you can work on code in a team but also means you could show off your best work on an online portfolio on GitLab or contribute to open source projects.

## Git’s structure

Git monitors all changes in a project but you choose **which changes are recorded and when**. 

Changes are organised into chronological **commits** made up of a set of changes. 

To record changes you must **stage** each modified file for the next commit. 

Process:
1. **Working directory** is your versioned folder of the code which Git monitors. 
2. **Staging area** is where changed files are added for the changes to be _recorded_ as a commit.
3. **Git directory** stores all the _committed_ changes with associated commit messages to describe changes.
4. **Remote repository** is a remote server which stores the committed changes you **push**. Colleagues can **pull** your code from the remote server.

For a project tracked with Git, there are multiple phases in the process of recording your changes. It is important to remember that just because you changed something in a file -- this change won't be automatically recorded. In particular, Git won't track **any** changes in a newly created file. This is because you might have files you don't want to track (e.g. sensitive passwords or large data files).

All your code lives in the **working directory** on your local machine. This is your version of the codebase and isn't directly affected by other people adding code on their versions or to the remote server. When you want to add a change in your code you **add** this to the **staging** area. You can add multiple files to the staging area and **every staged change** will form part of the next commit you create. Your project will have a folder called `.git` which is the **git directory** storing all the information about code versions. You usually won't have to touch this. Finally, we typically use Git with a remote server platform (e.g. GitLab) storing a **remote repository** which you **push** changes to or **pull** changes from. The remote version is usually what everyone in the team contributes towards and Git has a coordination system so these changes can all work together.

## Code as a commit history

Every commit is made up of a **patch** (all the lines added, removed or changes from relevant files) and **message** describing the changes e.g.
```
CURRENT_CODE = PREVIOUS_CODE_VERSION + YOUR_COMMIT
```
Every **commit** should have a single purpose e.g.
```
commit e6a79938cd72140f857c4828071fd41e0d994342                                                                   
Author: example_user <example@mail.com>                                                                     
Date:   Wed Sep 2 14:27:51 2020 +0100                                                                                 Update functional test 2.4 for Numpy>=1.0.3
```
If you use Git from the start, then every file has all changes tracked **starting from creating a blank file** and working up in small changes to the **current working version**.

Each commit in Git describes a set of **changes** to the project as a **patch** on the **previous version of the code**. This will usually be added and deleted lines in files but could also be files being renamed or deleted as well as complex project reconfigurations. A commit should **always** be accompanied by a commit **message** which describes what the changes in the commit does. In most workplaces, the commit message should also include an issue identifier related to the task being worked on, which enables the issue tracking software (e.g.: Jira) to tie code changes to required changes to the code. This means your project can always be managed as a series of code changes building on top of each other from the initial commit (which usually is just blank files). Having a history of your code can be very useful. You can go back to use old versions of different functions if something breaks or test new features before launching them to the user. This is all possible through Git.

We mentioned that every line or file that is added, deleted or changed counts as a change in Git. Now, if we create a new file and start tracking every change then we have the full history of this file. We can use Git on this file to see the current state as a series of incremental additions and changes starting from a blank file up to the current version. This means we can monitor how each file in our projects has changed over time.

## Cloning and authentication

Two options for cloning a repository:

| SSH | HTTPS |
|-|-|
| Protocol that uses encryption to secure the connection between a client and a server, originally created for secure terminal access. | Protocol that uses encryption to secure the connection between a client and a server, originally created for securely accessing webites over an insecure network. |
| Authentication works with a  **private** key and a   **public** key that needs to be generated. | Client provides **authentication information (username, password)** and server verifies information. |
| **The private key must be kept confidential**. The public key can be shared with the GitLab server. | Similar to any website login, **your password must be kept confidential**.  |

The process of creating your own version of the remote repository to work on is called "cloning". There are two protocols allowed for cloning this repository from GitLab -- SSH or HTTPS. They are subtly different and there are some occasions when you can **only use one protocol** for downloading a project. SSH requires generating public and private keys that are unique to your user. You give your **public** key to GitLab and when you try and clone the repository you authenticate with your related **private** key to confirm your identity and allow you to clone the project. Alternatively, you can clone using HTTPS - which requires your username and password for identification. Some people prefer to use SSH so you don't have to enter passwords all the time but note you should **never share your SSH private key**.

## Pulling and merging

The history of the project is a **series of commits**. Co-workers might add commits to the server that you need to integrate into your code to stay up to date. 


The opposite of **`git push`** is **`git pull`** which retrieves all the **new changes** from the remote server and applies each change to your code. 

**`git pull`** is actually made up of two commands:
- **`git fetch`**: retrieve new commits from the remote server but **store them without applying each patch**.
- **`git merge`**: integrate each new commit into **your local version** from oldest to newest. Your version should be **even** with the remote server after this.



The inverse command to **git push** is the **git pull** command. This command will retrieve all the new commits in the remote repository that **your local version doesn't have yet** and update your local codebase. This can happen when you merge code between branches on the remote repository (i.e. in a GitLab merge request) and then need to update your local version. This can also happen when you are working with someone on a branch and they push changes that you need to build on. 

Sometimes, you will have local changes that **conflict** with the remote changes that are trying to be merged. This is a **merge conflict**.

## Pulling and merging



Diagram from [this medium article on Fetch vs Pull.](https://medium.com/@sabbirhossain_70520/git-fetch-vs-git-pull-691823ed4239)

We can look at what we just discussed using a diagram. We saw three commands for some form of retrieval from the remote server: `fetch`, `merge` and `pull`. In this diagram, `origin` is the remote server and `local` is your own copy of the code on `main` branch. We want to get changes from `origin` and integrate these commits into our own code. The first step in this process is `git fetch` which will check for new commits in `origin`, download them and store them locally as `origin/main`. This process doesn't yet add these changes to our local copy. To integrate the changes we retrieved using `git fetch`, we use `git merge` to add them onto our own local codebase. Now our local copy matches the commits in `origin`. The command `git pull` conveniently **combines** these two commands into one step. As long as you don't have any conflicts, then you can usually run `git pull` to retrieve new remote commits. 

````python
%%bash

git pull
````

Since we or anyone else have not **push**ed any commits to remote repository, **git pull** reports back that everything is "Already up to date.". i.e. no new commits to fetch.

## Branches

Branches are **different versions of the same project.** The default branch is usually **master** or **main**.

A common workflow with branches:
  1. **`main`** is the stable version of the project with only tested and working features.
  2. You create a **new branch** from **`main`** to develop a new feature. 
  3. Commits for the feature are added to **only the new branch**.
  4. When the feature is complete you **merge** the branch to **`main`**.
  5. **`main`** now includes all the commits from the branch and your new feature.

So far we've seen the word `main` in various messages from Git. We can now explain what this means for your workflow. Git histories are organised into branches describing different versions of the same project as different commit histories. Branches allow us to work on multiple features in isolation on different branches and maintain separation between our own changes and our colleagues other work. 

You will typically always start off with a **single, default branch** called `main` which usually is used to maintain the stable and tested version of a code project. We can use branches to diverge from `main` to work on some new feature in a separate, new branch. This branch is **isolated** from main with a different history of **new commits** representing your new changes. When the feature is complete, we want to **merge** these changes back into `main` (more on this later). When this is merged, the commits from the branch are **added** to the `main` branch.

## Merging a branch

The `update_readme` branch has new changes we want in `main`.

We can merge them on **GitLab** in a **merge request**.

**First** push `update_readme` branch to the server:
```
$ git push --set-upstream origin update_readme
...
To https://gitlab.com/tomsherborne/cspark-demo.git
 * [new branch]      update_readme -> update_readme
Branch 'update_readme' set up to track remote branch 'update_readme' from 'origin'.
```


**Secondly** you create a **merge request** in **GitLab** to merge **update_readme** into **main**. 

Add details and a description then submit. If you are working in a team you would assign **reviewers** to review your code.

**Note:** On some other systems (e.g.: Atlassian BitBucket) the terminology for "Merge Request" is a "Pull Request"

````python
%%bash

git checkout update_readme
git push --set-upstream origin update_readme
````

After creating a new commit in the branch, we want to push these changes to the remote repository. The remote version of the project **also uses branches** and we want our local changes to exist in the same branch remotely. For this, we need `git push` to move our changes to the remote server. Because the server doesn't know which branches exist locally at first, we use the `--set-upstream-to origin` argument to tell the GitLab server to make a new branch with the same name. Following this, when we push to this branch, we can use `git push` alone. Our changes are now in the remote GitLab server. Now to merge the changes from this branch we open a **merge request** inside the GitLab web platform. ## Git jargon

- **Repository**: the online version of your git project also called the **remote**.
- **Fork**: a complete copy of someone else's repository for your own use. Your commits to a forked repository **do not** affect the original repository you copied. Changes in the original repository might affect your own, if you choose to enable fork syncing. 
- **Branch**: different versions of your repository.
- **Pull or Merge Request**: action to merge one branch into another by applying the commits from one branch to another.
- **Conflict**: simultaneous divergent changes to the same file will conflict.

Finally, let's go over some of the terminology used in Git which can quickly be confusing for new users. 

The **repository** refers to the remote server version of a code project. We created and modified a repository hosted on GitLab. 

Sometimes, someone else has code project hosted in a repository which you might want to copy or modify for your needs. You can create a **fork** of their project as a copy into your own user space to work on. Your **fork** of the existing project is mostly separate and you could make any changes you want. You won't receive any updates if they update their original code unless you specifically pull them from the **original** repository (more on this in the next lesson on Git) or enable fork syncing. 

A **branch** is a version of the project following a specific commit history. If you created a new branch from `main`, then the commit in `main` before you started adding `branch` specific changes is sometimes called where you _branched off_. 

A **pull or merge request** is the request to merge changes from one branch into another in an remote repository. This is where tests might be run to ensure working functionality and colleagues might review your changes and ask for more. 

A **conflict** is where two commits modify the same lines of code and Git can't decide which change to prioritise. These can be fixed manually in each file with conflicts.

### Summary

- Git is a version control system to **track changes** in a project and distribute them to a server and co-workers.
- The history of a project is a series of incremental **commits** describing code changes.
- New **commits** are created through **making** changes, **staging** changes then **committing** changes. 
- Different histories can be tracked in different **branches**.
- `main` is the default **branch** and tracks the stable version of the project. 
- Create new branches for new features then **merge** feature branches to `main` when complete.
